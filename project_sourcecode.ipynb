```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_m = pd.read_csv('mcdonalds_nutrition.csv') #Loading Mcdonalds nutrition data frame
df_s = pd.read_csv('starbucks_nutrition.csv') #Loading Starbucks nutrition data frame

#Dimensionality reduction

df_m1 = df_m[['Vendor','Item','Calories','Fat','Sodium','Carbs', 'Fiber','Protein']] 

options = ['Ginger Lemonade','Orange Juice','Apple Juice','Coffee','Hazelnut Iced Coffee', 'Hot Chocolate', 'Bacon Clubhouse Grilled Chicken Sandwich','Chicken BLT Salad Sandwich', 'Chocolate Chip Cookie', 'Oatmeal Raisin Cookie'] 
 

rslt_df = df_m1[df_m1['Item'].isin(options)]

df_s1 = df_s[['Vendor','Item','Calories','Fat','Sodium','Carbs', 'Fiber','Protein']]

rslt_df1 = df_s1[df_s1['Item'].isin(options)]

#print(rslt_df)
#print(rslt_df1)

#combining two data frames
pd.concat([rslt_df,rslt_df1])

#df_combine = pd.concat([rslt_df,rslt_df1])

#print (df_combine)

#Graph Chicken BLT Sandwich

data1= [450,15,1230,43,3,36]
data2= [470,25, 15,35,3,21]
labels = ['Calories','Fat','Sodium','Carbs','Fiber','Protein']
width= 0.3
plt.title('Chicken BLT Sandwich')
plt.xticks(range(len(data1)), labels)
plt.bar(np.arange(len(data1)), data1, width=width)
plt.bar(np.arange(len(data2))+ width, data2, width=width)
V = ['McDonalds','Starbucks']
plt.legend(V,loc=2)
plt.show()

#Graph Bacon Clubhouse Grilled Chicken Sandwich
data1= [590,25,1560,51,4,40]
data2= [370,19, 220,31,1,18]
plt.title('Bacon Cluhouse Grilled Chicken Sandwich')
plt.xticks(range(len(data1)), labels)
plt.bar(np.arange(len(data1)), data1, width=width)
plt.bar(np.arange(len(data2))+ width, data2, width=width)
V = ['McDonalds','Starbucks']
plt.legend(V,loc=2)
plt.show()

#Graph Chocolate Chip Cookie
data1= [160,8,9,21,1,2]
data2= [310,15, 150,42,2,4]
plt.title('Chocolate Chip Cookie')
plt.xticks(range(len(data1)), labels)
plt.bar(np.arange(len(data1)), data1, width=width)
plt.bar(np.arange(len(data2))+ width, data2, width=width)
V = ['McDonalds','Starbucks']
plt.legend(V,loc=2)
plt.show()


#Graph Hot Chocolate
data1= [360,13,180,50,1,11]
data2= [320,9, 160,47,4,14]
plt.title('Hot Chocolate')
plt.xticks(range(len(data1)), labels)
plt.bar(np.arange(len(data1)), data1, width=width)
plt.bar(np.arange(len(data2))+ width, data2, width=width)
V = ['McDonalds','Starbucks']
plt.legend(V,loc=2)
plt.show()

```
